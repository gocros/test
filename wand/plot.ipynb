```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('/tmp/capacity.txt')

fig,ax1=plt.subplots(figsize=(16,4))

ax2 = ax1.twinx()

ax1.plot(df["c0"], 'r', label='Capacity0')
ax1.plot(df["c1"], 'b', label='Capacity1')
ax2.plot(df["v0"]/1e6, 'r--', label='Voltage0')
ax2.plot(df["v1"]/1e6, 'b--', label='Voltage1')
ax2.set_ylim([6,9])

fig.legend(loc='right')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('/tmp/capacity.txt')

fig,ax1=plt.subplots(figsize=(16,4))

ax2 = ax1.twinx()

ax1.plot(df["i0"]/1e6, 'r', label='Current0')
ax1.plot(df["i1"]/1e6, 'r--', label='Current1')
ax2.plot(df["v0"]/1e6, 'b', label='Voltage0')
ax2.plot(df["v1"]/1e6, 'b--', label='Voltage1')
ax2.set_ylim([6,9])
ax1.set_ylabel('Current (A)', color='r')
ax2.set_ylabel('Voltage (V)', color='b')
```
